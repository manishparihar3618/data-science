# API (Application Programming Interface)

## Definition
An *API (Application Programming Interface)* is a set of rules that allows two software applications to communicate with each other and exchange data.

Think of an API as a *messenger* between the client and the server.

---

## Real-World Example (Train Ticket Booking)

Suppose we want to book a train ticket.

There are multiple platforms:

- IRCTC
- MakeMyTrip
- Yatra.com

Imagine someone books Seat *B142* on the *Agra Express* through the IRCTC website.

Now, if another user searches for the same seat on MakeMyTrip or Yatra, the seat should immediately appear as *Booked* or *Not Available*.

This synchronization happens through *APIs*.

Each website has its own application and database, but they communicate using APIs to exchange information.

---

## Important Concept

Different websites may have different databases.

MakeMyTrip does *not* directly access IRCTC's database.

Instead:

1. MakeMyTrip sends an API request.
2. The IRCTC server processes the request.
3. IRCTC checks its database.
4. The server returns the required information.
5. MakeMyTrip displays the updated result to the user.

## API Does NOT Directly Access the Database

An API itself does *not* access the database.

The application sends a request to the server.

The server may:
- Read data from the database
- Process the request
- Perform authentication
- Apply business logic
- Return the response

---

## API Returns Data

Most APIs return data in *JSON (JavaScript Object Notation)* format.

Example:

json
{
    "train": "Agra Express",
    "seat": "B142",
    "status": "Booked"
}

## Summary

An API is a communication bridge between two applications.

It follows predefined rules for exchanging data.

Applications do not directly communicate with databases; they communicate with servers through APIs, and servers return the required information.
