## Model Training (Logistic)

```python
X = df.drop("TenYearCHD", axis=1)
y = df["TenYearCHD"]
```

train-test-split 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,y_train,X_test,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
# Standard Scaler 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
num_cols = [
    "age",
    "cigsPerDay",
    "totChol",
    "sysBP",
    "diaBP",
    "BMI",
    "heartRate",
    "glucose"
]

binary_cols = [
    "male",
    "education",
    "currentSmoker",
    "BPMeds",
    "prevalentStroke",
    "prevalentHyp",
    "diabetes"
]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

train

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(class_weight="balanced")
logistic.fit(X_train,y_train)
```

```python
y_pred = logistic.predict(X_test)
```

performance metrics 

```python
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report
)



print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

Roc_Auc

```python
y_prob = logistic.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import roc_auc_score
print("Roc-auc:", roc_auc_score(y_test, y_prob))
```

Hyperparameter Tuning

GridSearchCV

```python
model = LogisticRegression(class_weight="balanced",random_state=42)
penalty = ['l1', 'l2', 'elasticnet']
c_values = [100,10,1.0,0.1,0.01]
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params = dict(penalty = penalty, C=c_values, solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
grid_model = GridSearchCV(cv=cv,estimator=model,param_grid=params,scoring='f1',n_jobs=-1)
# using f1 in place of accuracy but still same identical perfornmance of model
```

```python
grid_model
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

```python
y_pred=grid_model.predict(X_test)
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
