## EDA and Data Cleaning on Algerian_forest_fires_dataset.csv


```python
import numpy as np 
import pandas as pd
import matplotlib as plt 
import seaborn as sns 
```

## Data Cleaning

```python
df[df.isnull().any(axis=1)]
# Sidi bel region 
```

```python
df.loc[:122,"Region"]=0 
df.loc[122:,"Region"]=1

# sidibel region = 0
# begial region = 1
```

```python
df.info()
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
# Region become int 
```

```python
df.isnull().sum() #Total null value 
```

```python
df=df.dropna().reset_index(drop=True) # Dropping null values 
df.head()
```

```python
df.isnull().sum() # null values removed
```

```python
df.iloc[[122]] # should drop this row 
```

```python
df=df.drop(122).reset_index(drop=True) #Dropped the 122nd row 
```

```python
df.columns
#There ae spaces in column names like space after Classes
```

```python
## Fix spaces in columns names
df.columns = df.columns.str.strip() #strip() method remove widespaces 
df.columns
```

```python
df.info()
```

### Change the required columns as integer data type

```python
df.columns 
```

```python
df[['month','day','year','Temperature','RH','Ws']]= df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

### Changing the other columns to float data types 

```python
str=[features for features in df.columns if df[features].dtypes=='str']
```

```python
for i in str:
    if i!= 'Classes' :
     df[i]=df[i].astype(float)
```

```python
df[["Classes"]] = df[["Classes"]].astype(object)
```

```python
df.info()
```

```python
df.describe()
```

### Lets save the cleaned Dataset

```python
df.to_csv('Algerian_forest_fire_cleaned_dataset.csv',index = False)
```
