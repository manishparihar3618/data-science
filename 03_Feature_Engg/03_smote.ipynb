# SMOTE(Synthetic Minority Oversampling Technique)

SMOTE is a technique used in ML to address imbalaced datasets where the minority class has significantly fewer isntances than the majority class. SMOTE involves generating synthetic instances of minority class by interpolating between existing instances.

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=12,)
```

```python
import pandas as pd
# Create a 90:10 imbalanced classification dataset
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df['target'].value_counts() #Verify class imbalace
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c =final_df['target']) #visualize
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# transform the dataset using Smote
oversample=SMOTE()
x,y=oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
```

```python
plt.scatter(oversample_df['f1'],oversample_df['f2'],c=oversample_df['target'])
#After SMOTE both classes become equal 
```

Imbalanced Dataset = One class has significantly more samples than another.

Upsampling  = Duplicate minority class samples.

Downsampling = Remove majority class samples.

SMOTE = Generate synthetic minority samples using nearest neighbors.

Upsampling Risk  = Overfitting (duplicate records).
Downsampling Risk = Information loss.
SMOTE Risk = May create noisy synthetic samples.

Rule:
 Upsampling copies data, SMOTE creates new data.
