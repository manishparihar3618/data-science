## Handling a Imbalanced Dataset

```python
import numpy as np
import pandas as pd

# here we are createing a imbalaced dataset 
np.random.seed(123) #Fix random seed for reproduciblity
n_sample = 1000
class_0_ratio = 0.9
n_class_0 = int(n_sample * class_0_ratio)
n_class_1 = n_sample - n_class_0
```

```python
n_class_0, n_class_1
```

### Create Dataframe with imbalanced dataset

```python
class_0 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2' : np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
    }) # Majority class (Target = 0)


class_1 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature_2' : np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1]*n_class_1
    }) # Minority class (target = 1)
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True) #Combine both class
```

```python
df.head() 
```

```python
df.tail()
```

```python
df['target'].value_counts() # Verify 
```

### (1) UP SAMPLING

```python
#Seprate majority and minority classes
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python

from sklearn.utils import resample
# Upsample minority class to match majority class
# replace=True allow repeatation sampling
df_minority_upsampled=resample(df_minority,replace=True, #Sample with replacement 
n_samples=len(df_majority),random_state=42)
```

```python
df_minority_upsampled.shape #new class cheak
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled['target'].value_counts()
```

### (2) Down Sampling

```python
import numpy as np
import pandas as pd

np.random.seed(123)
n_sample = 1000
class_0_ratio = 0.9
n_class_0 = int(n_sample * class_0_ratio)
n_class_1 = n_sample - n_class_0


class_0 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2' : np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
    })


class_1 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature_2' : np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1]*n_class_1
    })


df=pd.concat([class_0,class_1]).reset_index(drop=True)

df['target'].value_counts() #Cheak the class distribution
```

```python
#downsampling
#Separate majority and minority classes 
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python

from sklearn.utils import resample
#downsample the majority class to match minor class size
#replace=false means no duplication sampling
df_majority_downsampled=resample(df_majority,replace=False, 
n_samples=len(df_minority),random_state=42)
```

```python
df_majority_downsampled.shape # reduced to 100 from 900 to match df_minority
```

```python
df_downsampled = pd.concat([df_majority_downsampled,df_minority]) #merge 
```

```python
df_downsampled['target'].value_counts() #Downsampling is bad because we are lossing datapoints
```

Revision:

Upsampling  = Increase minority class (duplicate samples).

Downsampling =  Reduce majority class (remove samples).

Upsampling risk = Overfitting.

Downsampling risk = Information loss.
