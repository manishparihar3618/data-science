### Nominal or One Hot Encoding

This technique is used to represent categorical data as numerical data.

disdvantages  - we have 100 categories it will create 100 new features
             
              - Sparse matrix (1s,0s) -> cause overfitting

```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
```

```python
## Create a simple dataframe
df= pd.DataFrame({
        'color' : ['red','blue','green','green','red','blue']
    })
```

```python
df.head()
```

```python
## Create an instance of OneHotencoder
encoder= OneHotEncoder()
```

```python
#Perform fit and tranform 
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd # Convert into numerical format
encoder_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
encoder.transform([['blue']]).toarray() #transform new data
```

```python
pd.concat([df,encoder_df],axis=1) #merge with original data
```

### Ordinal Encoding

It is used to encode categorical data in ranks. In this technique, each category is assigned a numerical value based on its position in the order. For example, if we have categorical variable "education level" with four possible values 

High school : 1

College : 2

Graduate : 3

post-graduate : 4

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    'size': ['small','medium', 'large', 'medium', 'small', 'large']
})
```

```python
df
```

```python
#create a instance of OrdinalEncoder then perform fit_tranform
encoder = OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```

### Target Guided Ordinal Encoding
It is a technique used to encode categorical variables based on thier relationship with the target variable. This encoding technique is useful when we have a categorical variable with large number of unique categories, and we want to ise this variable as feature in machine learning model.

In This Encoding, we replace eacvh category in the categorical variable with numerical value based on mean and median of the target variable for that category.

```python
df = pd.DataFrame({
    'city':['New york','Londan','Paris','Tokyo','New york','Paris'],
    'price' : [200,150,300,250,180,320]
})
df
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```

```python
df[['city','price','city_encoded']]
```
